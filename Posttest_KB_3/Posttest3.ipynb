### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Membaca Dataset

```python
df = pd.read_csv('life expectancy.csv')
```

### Cek Data teratas

```python
df.head(10)
```

## **Pie Chart**

#### ~ Pie Chart dibawah menjelaskan tentang jumlah pendapatan grub, untuk persentase tertinggi disini dipegang oleh High Income sedangkan persentase terendah adalah low income jadi bisa disimpulkan bahwa jumlah pendapatan grub selama ini selalu meranjak high income alis untung banyak dibanding rugi.

```python
plt.figure(figsize=(15,10))

plt.pie(df.IncomeGroup.value_counts(), labels=df.IncomeGroup.value_counts().index, autopct="%1.1f%%", startangle=90, explode=[0.05,0,0,0])

plt.title("Jumlah Pendapatan grub")
plt.show()
```

## **HeatMap**

#### ~ Berdasarkan heatmap dibawah ini menjelaskan betapa tingginya/banyaknya suatu kelompok berdasarkan warna yang semakin menebal atau ditandakan dengan jumlah sempurna yaitu 1. Untuk banyaknya suatu kelompok disini dia memiliki nilai yang diagonal untuk yang sempurna dan memiliki nilai paling rendah di angka 0.006.

```python
plt.figure(figsize=(15,10))

sns.heatmap (
    df.select_dtypes(include=['float']).corr(),
    annot = True,
    cmap = 'ocean',
)
plt.show()
```

## **Box Plot**

#### ~ Pada Boxplot dibawah, Menjelaskan data kekurangan gizi per tahunnya. Dibawah ini terdapat pada tahun 2001 memiliki data tertinggi hingga 20+ sedangkan data terendah terjadi pada tahun 2015 hanya 13 orang saja. Selain itu banyak sekali data disini yang agak lain(outliers) yang keluar dari data yang terterah, berarti masih memiliki data yang cukup menyimpang dari data tersebut

```python
plt.figure(figsize=(20,15))

sns.boxplot (df, y = "Prevelance of Undernourishment", x = "Year",)

plt.title("Kekurangan Gizi per tahun", fontweight="bold", fontsize="20")
plt.show()
```

## **HISTOGRAM**

#### ~ Pada tabel Histogram dibawah, penyakit kekurangan gizi yang memiliki frekuensi tertinggi terjadi pada usia balita, karena mungkin masih banyak orang tua yang belum sadar betapa penting gizi sebagai pertumbuhan balita dan anak-anak.

```python
plt.figure(figsize=(15,10))

plt.hist(df['Prevelance of Undernourishment'], bins=20, color='skyblue')
plt.title('Penyakit Kekurangan Gizi', fontweight="bold")
plt.ylabel('Frekuensi Kekurangan Gizi')
plt.xlabel('Usia')
plt.show()
```
