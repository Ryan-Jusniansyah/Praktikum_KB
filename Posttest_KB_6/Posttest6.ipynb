### Import Library

```python
import pandas as pd
import numpy as np
```

### Mengakses data CSV

```python
df = pd.read_csv('life expectancy.csv')
```

### Copy database

```python
df2 = df.copy()
df2
```

### Menghapus data yang memiliki nilai NaN

```python
df2.isna().any()
```

```python
df2.dropna(inplace=True)
df2.isna().any()
```

### Melakukan drop columns pada kolom yang tidak digunakan

```python
df2
```

```python
df2.drop(columns=['Country Name', 'Country Code', 'Region', 'Year'], inplace=True)
df2
```

### Pemilihan Feature dan target

```python
X = df2.drop(columns=['IncomeGroup'])
y = df2['IncomeGroup']
y
```

### Melakukan normalisasi data dengan MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

columns = list(X.columns)
X[columns] = mms.fit_transform(X[columns])
```

### Melakukan Label encoder pada IncomeGroup

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
```

```python
y = l.fit_transform(y)
y
```

### Melakukan Split data

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3)

print(f'panjang X_train = {len(X_train)}')
print(f'panjang X_test = {len(X_test)}')
print(f'panjang y_train = {len(y_train)}')
print(f'panjang y_test = {len(y_test)}')
```

<p>diatas untuk mengecek panjang data diantar train dan test</p>

### Modeling

```python
from keras import Sequential as seq
from keras.layers import Dense as De
```

```python
model = seq()

#input layer
model.add(De(8, activation='relu', input_dim=11))

# Hidden Layer
model.add(De(16, activation='relu'))
model.add(De(32, activation='relu'))

# Output Layer
model.add(De(1, activation='sigmoid'))
```

```python
model.summary()
```

<p>Label diatas menunjukkan penggunaan neuron pada input, hidden, dan output layer.</p>

### Memanggil fungsi callback

```python
from keras.callbacks import EarlyStopping

monitor_callbacks = EarlyStopping(monitor='val_accuracy', patience=10)
```

### Model Compile

```python
from keras.optimizers import Adam

model.compile(
    optimizer = 'Adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

### Training Model

```python
Train = model.fit(
    x = X_train,
    y = y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    verbose=1,
    callbacks=[monitor_callbacks]
)
```

### Model Evaluasi

```python
model.evaluate(X_test, y_test)
```

<p>pada loss sendiri disini terdapat 0.5129 loss ini cukup besar maka dari itu mesti dilatih lagi datanya agar mencapai sempurna karena jika loss semakin rendah maka data semakin mudah digunakan.</p>

<p>Sedangkan accuracy sendiri mencapai 0.5348 lebih tinggi dibandingkan lossnya tapi jika accuracy ini bisa lebih tinggi lagi, maka data sudah dapat dinyatakan bagus</p>

### Membuat table untuk loss dan juga evaluasi

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
plt.subplot(1, 2, 1)
plt.plot(Train.history['accuracy'], label='Training Accuracy')
plt.plot(Train.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='upper right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(Train.history['loss'], label='Training loss')
plt.plot(Train.history['val_loss'], label='Validation loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.show()
```
