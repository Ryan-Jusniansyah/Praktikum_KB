### Import Library

```python
import pandas as pd
import numpy as np
```

### Mengakses data CSV

```python
df = pd.read_csv('life expectancy.csv')
```

### Copy database

```python
df2 = df.copy()
df2
```

### Menghapus data yang memiliki nilai NaN

```python
df2.isna().any()
```

```python
df2.dropna(inplace=True)
df2.isna().any()
```

### Melakukan drop columns pada kolom yang tidak digunakan

```python
df2
```

```python
df2.drop(columns=['Country Name', 'Country Code', 'Region', 'Year'], inplace=True)
df2
```

### Pemilihan Feature dan target

```python
X = df2.drop(columns=['IncomeGroup'])
y = df2['IncomeGroup']
y
```

### Melakukan normalisasi data dengan MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

columns = list(X.columns)
X[columns] = mms.fit_transform(X[columns])
```

### Melakukan Label encoder pada IncomeGroup

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
```

```python
y = l.fit_transform(y)
y
```

### Melakukan Split data

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3)

print(f'panjang X_train = {len(X_train)}')
print(f'panjang X_test = {len(X_test)}')
print(f'panjang y_train = {len(y_train)}')
print(f'panjang y_test = {len(y_test)}')
```

### Modeling

```python
from keras import Sequential as seq
from keras.layers import Dense as De
```

```python
model = seq()

#input layer
model.add(De(8, activation='relu', input_dim=11))

# Hidden Layer
model.add(De(16, activation='relu'))
model.add(De(32, activation='relu'))

# Output Layer
model.add(De(1, activation='sigmoid'))
```

### Memanggil fungsi callback

```python
from keras.callbacks import EarlyStopping

monitor_callbacks = EarlyStopping(monitor='val_accuracy', patience=10)
```

### Model Compile

```python
from keras.optimizers import Adam

model.compile(
    optimizer = 'Adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

### Training Model

```python
model.fit(
    x = X_train,
    y = y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    verbose=1,
    callbacks=[monitor_callbacks]
)
```
