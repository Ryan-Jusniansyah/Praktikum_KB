### ~ Tampilkan Informasi dalam bentuk Dictionary

```python
Dtypes = df.select_dtypes(include=['int','float'])
jumlah = {
    "jumlah_record" : df.count(),
    "jumlah_attribute" : len(df.columns),
    "nama_attribute" : df.columns,
    "jumlah_tipeAngka" : len(Dtypes.columns)
}

print(jumlah["jumlah_record"])
print("="*30)
print(jumlah["jumlah_attribute"])
print("="*30)
print(jumlah["nama_attribute"])
print("="*30)
print(jumlah["jumlah_tipeAngka"])
```
