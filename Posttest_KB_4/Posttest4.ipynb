### ~ One Hot Encoding dengan dataset yang saya miliki

```python
from sklearn.preprocessing import OneHotEncoder
```

##### Deklarasi One Hot Encoder

```python
OHE = OneHotEncoder(sparse_output=False)
```

##### Mengecek unique pada attribute income groub

```python
df.IncomeGroup.unique()
```

##### Menambahkan sebuah kolom baru dengan nilai 1 dan 0 yang menandakan di table tersebut sesuai atau tidak, jika ada dia akan menilai 1, jika tidak maka akan memunculkan 0

```python
Income = OHE.fit_transform(df[["IncomeGroup"]])
Income = pd.DataFrame(Income)

df['Low income'] = Income[0]
df['Lower middle income'] = Income[1]
df['Upper middle income'] = Income[2]
df['High income'] = Income[3]
```
